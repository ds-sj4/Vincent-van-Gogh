```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
art = pd.read_csv("art_ratings.csv", encoding = 'UTF-8')
art['year'] = art['year'].astype(str)
art = art.loc[ art['life'] != "Beginning" ]
art.head()
```

```python
plot = sns.pairplot(art, hue = 'group')
plot
```

```python
plot.fig.savefig('Scatter.png', dpi=300, bbox_inches='tight')
```

```python
p1 = sns.lmplot(data = art, x = 'valence', y = 'creativity', col = 'life')
p1
```

```python
p2 = sns.lmplot(data = art, x = 'arousal', y = 'creativity', col = 'life')
p2
```

```python
p3 = sns.lmplot(data = art, x = 'cohesivity', y = 'creativity', col = 'life')
p3
```

```python
p1.fig.savefig('Valence.png', dpi=300, bbox_inches='tight')
p2.fig.savefig('Arousal.png', dpi=300, bbox_inches='tight')
p3.fig.savefig('Cohesivity.png', dpi=300, bbox_inches='tight')
```
